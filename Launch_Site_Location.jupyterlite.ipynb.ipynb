```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
import requests
import io
from IPython.display import display

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'

# Fetch CSV file using requests
response = requests.get(URL)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Create DataFrame from CSV content
    spacex_csv_file = io.BytesIO(response.content)
    spacex_df = pd.read_csv(spacex_csv_file)

    # Create a folium map centered around a location (you can adjust the location and zoom level)
    map_center = [28.7041, 77.1025]  # Latitude and longitude of a central location
    spacex_map = folium.Map(location=map_center, zoom_start=2)

    # Create a MarkerCluster
    marker_cluster = MarkerCluster().add_to(spacex_map)

    # Iterate through the launch sites and add markers to the map
    for index, site in spacex_df.iterrows():
        # Customize the popup content as needed
        popup_content = f"Launch Site: {site['Launch Site']}<br>Location: {site['Location'] if 'Location' in spacex_df.columns else 'N/A'}<br>Status: {site['class']}"

        # Set marker color based on success or failure
        marker_color = 'green' if site['class'] == 1 else 'red'

        # Add a marker to the MarkerCluster
        folium.Marker(
            location=[site['Lat'], site['Long']],
            popup=folium.Popup(popup_content, parse_html=True),
            icon=folium.Icon(color=marker_color, icon='rocket', prefix='fa')  # Custom icon using Font Awesome
        ).add_to(marker_cluster)

    # Display the map
    display(spacex_map)
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from geopy.distance import geodesic
import pandas as pd
import requests
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'

# Fetch CSV file using requests
response = requests.get(URL)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Create DataFrame from CSV content
    spacex_csv_file = io.BytesIO(response.content)
    spacex_df = pd.read_csv(spacex_csv_file)

    # Example: Adding columns 'Closest_lat' and 'Closest_long' with sample proximity coordinates
    spacex_df['Closest_lat'] = 28.562301
    spacex_df['Closest_long'] = 77.263203

    # Function to calculate distance using Haversine formula
    def calculate_distance(row):
        launch_site_coords = (row['Lat'], row['Long'])
        proximity_coords = (row['Closest_lat'], row['Closest_long'])
        return geodesic(launch_site_coords, proximity_coords).kilometers

    # Calculate distances using the calculate_distance function
    spacex_df['Distance_to_proximity'] = spacex_df.apply(calculate_distance, axis=1)

    # Display the DataFrame with distance information
    print(spacex_df[['Launch Site', 'Distance_to_proximity']])
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```
